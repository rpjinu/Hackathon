```python
# Mapping function for 'emp_length' column to convert to float
def convert_emp_length(emp_length):
    if emp_length == '10+ years':
        return 10.0
    elif emp_length == '< 1 year':
        return 0.0
    elif isinstance(emp_length, str):  # Handles cases like '3 years', '1 year', etc.
        return float(emp_length.split()[0])
    else:
        return np.nan  # Use np.nan for missing values

# Apply the function to both train and test datasets
train_data['emp_length'] = train_data['emp_length'].apply(convert_emp_length)
test_data['emp_length'] = test_data['emp_length'].apply(convert_emp_length)
```

```python
# Fill missing values with median for numerical columns
numerical_columns = ['emp_length', 'num_actv_bc_tl', 'mort_acc', 'tot_cur_bal', 'pub_rec_bankruptcies', 'revol_util']
for col in numerical_columns:
    train_data[col].fillna(train_data[col].median(), inplace=True)
    test_data[col].fillna(test_data[col].median(), inplace=True)
```

```python
# Fill missing values with mode for categorical columns
categorical_columns = ['emp_title', 'title']
for col in categorical_columns:
    train_data[col].fillna(train_data[col].mode()[0], inplace=True)
    test_data[col].fillna(test_data[col].mode()[0], inplace=True)
```

```python
#check the null values
print("check train dataset null values:-",train_data.isnull().sum().any())
print("check null values in test data:-",test_data.isnull().sum().any())
```

```python
#check the numerical column list
numerical_columns = train_data.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
# Convert the specified numerical columns to float64
train_data[numerical_columns] = train_data[numerical_columns].astype('float64')
test_data[numerical_columns] = test_data[numerical_columns].astype('float64')
```

```python
#drop date column
train_data.drop('earliest_cr_line', axis=1, inplace=True)
test_data.drop('earliest_cr_line', axis=1, inplace=True)
```

```python
# Apply the mapping to the loan_status column
train_data['loan_status'] = train_data['loan_status'].map({'Paid': 1,'Defaulted': 0})
```

```python
#title and porpose of loan are same so we drop title column to reduce dimensonality
train_data.drop('title', axis=1, inplace=True)
test_data.drop('title', axis=1, inplace=True)
```

```python
# Drop `emp_title` if it's not necessary
train_data = train_data.drop('emp_title', axis=1)
test_data = test_data.drop('emp_title', axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Define columns to encode
cols_to_encode = ['grade', 'initial_list_status', 'application_type', 'addr_state', 'sub_grade','home_ownership', 'purpose', 'term', 'verification_status']
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Encode train and test data separately
encoded_train = encoder.fit_transform(train_data[cols_to_encode])
encoded_test = encoder.transform(test_data[cols_to_encode])

# Convert to DataFrames with feature names
encoded_train_df = pd.DataFrame(encoded_train, columns=encoder.get_feature_names_out(cols_to_encode), index=train_data.index)
encoded_test_df = pd.DataFrame(encoded_test, columns=encoder.get_feature_names_out(cols_to_encode), index=test_data.index)

# Drop original columns and add encoded columns
train_data = train_data.drop(cols_to_encode, axis=1).join(encoded_train_df)
test_data = test_data.drop(cols_to_encode, axis=1).join(encoded_test_df)

# Align columns to ensure consistency
train_data, test_data = train_data.align(test_data, fill_value=0, axis=1)
```

```python
#drop the loan_staus column in test dataset
test_data = test_data.drop('loan_status', axis=1)
```

```python
# Step 1: Separate features and target in training data
X_train = train_data.drop(columns=['loan_status'])
y_train = train_data['loan_status']
```

```python
# Step 2: Train a classifier (Random Forest in this example)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict loan_status for the test data
test_predictions = model.predict(test_data)
```

***check accuracy there are no validaation dataset so we split train data to validation dataset ***

```python
# Check accuracy on a validation set (split from train_data)
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
model.fit(X_train_split, y_train_split)
val_predictions = model.predict(X_val_split)
print("Validation Accuracy:", accuracy_score(y_val_split, val_predictions))
print(classification_report(y_val_split, val_predictions))
```
